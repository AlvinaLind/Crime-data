```python
def get_row(df, keyword):
    # hitta första kolumnen dynamiskt
    first_col = df.columns[0]

    # gör första kolumnen till index
    df = df.set_index(first_col)

    # sök efter rad där indexet innehåller keyword
    return df[df.index.str.contains(keyword, na=False)]
result = get_row(df_151721, 'Rån inkl. grovt')
result
```
